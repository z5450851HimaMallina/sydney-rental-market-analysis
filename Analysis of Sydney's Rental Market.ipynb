```python


# Define weights (you may adjust these based on your judgment)
weights = {
    'Distance from UNSW': -1.0,  # Negative because lower is better
    'Median Weekly Rent': -0.5,
    'Weekly Transportation Cost': -0.3,
    'Number of Store': 0.2
}

# Normalize data
for column, weight in weights.items():
    max_value = data[column].max()
    min_value = data[column].min()
    # If weight is negative, we invert the normalized scale so higher original values result in lower normalized values
    if weight < 0:
        data[f'norm_{column}'] = 1 - ((data[column] - min_value) / (max_value - min_value))
    else:
        data[f'norm_{column}'] = (data[column] - min_value) / (max_value - min_value)

# Calculate weighted scores
data['score'] = sum(data[f'norm_{column}'] * abs(weight) for column, weight in weights.items())

# Identify the best suburb in each LGA
best_suburbs = data.loc[data.groupby('LGA')['score'].idxmax()]

# Identify the best LGA based on average score
best_lga = best_suburbs.loc[best_suburbs['score'].idxmax()]['LGA']

print("Best LGA:", best_lga)
print("Best Suburbs in each LGA:")
print(best_suburbs[['LGA', 'Suburb', 'score']])
```
